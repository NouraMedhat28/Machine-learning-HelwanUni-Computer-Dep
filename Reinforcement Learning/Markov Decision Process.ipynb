a  Markov decision Process is defined as the following :

State : where you are  in the environment

action : what decision does the agent takes

Transition function T(s,a,s') : the probability of the uncertainty in each state , in simpler word if you are state S , and took action A , with a probability T(s,a,s') you are going to go to s'

Reward(s,a,s') : the reward you get for taking an action

Discount Factor : caring about the present more than the future , by discounting the future by a discount rate 



# More in depth we defined :
Policy : a Mapping from State --> Actions , at state S what do you take? denoted by pi

Utility : which is the sum of discounted rewards following a speicifc policy.


Value of a Policy : the expected Utility of that policy , the expected discounted sum of rewards dentoed by Vpi and calculated by multiplying the Transition function by the utility.



Q value : after taking an action what's the expected utility following a speicif policy?
